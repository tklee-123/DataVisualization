```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
transport_percentage = pd.read_csv("data/the_problem/transport_emission.csv")
```

```python
colors = sns.color_palette("Set2", len(transport_percentage))


def autopct_format(pct):
    return ('%1.1f%%' % pct) if pct > 2 else ''  

# Plot
plt.figure(figsize=(10, 7))
wedges, texts, autotexts = plt.pie(
    transport_percentage["Percentage"],
    labels=None,  
    autopct=autopct_format,  
    startangle=140,
    colors=colors,
    pctdistance=0.85,  
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.2}
)


plt.legend(wedges, transport_percentage["Entity"],
           title="Loại phương tiện",
           loc="center left",
           bbox_to_anchor=(1, 0, 0.5, 1))  

# Tiêu đề và tỉ lệ trục
plt.title('Phân bố Mức độ phát thải khí ô nhiễm của các loại phương tiện', fontsize=14, weight='bold')
plt.axis('equal')  # Đảm bảo biểu đồ là hình tròn
plt.tight_layout()  # Điều chỉnh bố cục để tránh bị cắt

# Hiển thị biểu đồ
plt.show()
```
