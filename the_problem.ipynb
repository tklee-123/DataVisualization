```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
transport_percentage = pd.read_csv("data/the_problem/transport_emission.csv")
```

```python
colors = sns.color_palette("Set2", len(transport_percentage))


def autopct_format(pct):
    return ('%1.1f%%' % pct) if pct > 2 else ''  

# Plot
plt.figure(figsize=(10, 7))
wedges, texts, autotexts = plt.pie(
    transport_percentage["Percentage"],
    labels=None,  
    autopct=autopct_format,  
    startangle=140,
    colors=colors,
    pctdistance=0.85,  
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.2}
)


plt.legend(wedges, transport_percentage["Entity"],
           title="Loi phng tin",
           loc="center left",
           bbox_to_anchor=(1, 0, 0.5, 1))  

# Tiu  v t l trc
plt.title('Phn b Mc  pht thi kh  nhim ca cc loi phng tin', fontsize=14, weight='bold')
plt.axis('equal')  # m bo biu  l hnh trn
plt.tight_layout()  # iu chnh b cc  trnh b ct

# Hin th biu 
plt.show()
```
