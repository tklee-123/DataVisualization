```python
import pandas as pd

# Đọc dữ liệu từ file CSV với dấu phân cách là dấu chấm phẩy
file_path = 'data\\dioxide.csv'
data = pd.read_csv(file_path, delimiter=';')

# Hiển thị vài dòng đầu của dữ liệu để kiểm tra
print(data.head())
```

```python
import matplotlib.pyplot as plt

# Chuẩn bị dữ liệu cho biểu đồ cột
years = data.columns[1:]  # Lấy các cột đại diện cho các năm từ 1990 trở đi

# Vẽ biểu đồ cột cho từng loại phương tiện
plt.figure(figsize=(12, 8))

for index, row in data.iterrows():
    emissions = row.values[1:]  # Lấy lượng phát thải từ cột 1990 trở đi
    plt.plot(years, emissions, label=row['Category'])  # Vẽ đường biểu diễn cho mỗi loại

# Thiết lập các thông số cho biểu đồ
plt.xlabel('Năm')
plt.ylabel('Lượng phát thải')
plt.title('Xu hướng lượng phát thải qua các năm theo từng loại phương tiện')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Đưa chú thích ra ngoài đồ thị
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Bỏ dòng có tên "Total" nếu tồn tại trong cột 'Category'
data_filtered = data[data['Category'] != 'Total']

# Chuẩn bị dữ liệu cho năm 2000 và 2016 sau khi bỏ dòng "Total"
emissions_2000 = data_filtered['2000']
emissions_2016 = data_filtered['2016']
categories = data_filtered['Category']

# Tạo figure và hai biểu đồ tròn cạnh nhau
fig, axs = plt.subplots(1, 2, figsize=(14, 7))

# Biểu đồ tròn cho năm 2000
wedges_2000, texts_2000, autotexts_2000 = axs[0].pie(emissions_2000, labels=categories, autopct='%1.1f%%', startangle=140)
axs[0].set_title('Tỷ lệ phát thải năm 2000')

# Biểu đồ tròn cho năm 2016
wedges_2016, texts_2016, autotexts_2016 = axs[1].pie(emissions_2016, labels=categories, autopct='%1.1f%%', startangle=140)
axs[1].set_title('Tỷ lệ phát thải năm 2016')

# Thêm chú thích
for autotext in autotexts_2000 + autotexts_2016:
    autotext.set_color('white')
    autotext.set_fontsize(10)

# Thêm chú thích bên ngoài biểu đồ tròn
fig.legend(wedges_2000, categories, title="Phương tiện", loc="center left", bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load the CSV file to inspect its structure and columns
file_path = 'data\\VNGasolineCrawlData1020.csv'
data = pd.read_csv(file_path)

# Display the first few rows and column information to understand the dataset
data_info = data.info()
data_head = data.head()

data_info, data_head
```

```python
import matplotlib.pyplot as plt

# Filter out columns that contain mostly NaN values
filtered_data = data.drop(columns=['FO 3%S', 'FO 3,5%S', 'FO 3%S Price difference', 'FO 3,5%S Price difference'])

# Convert the 'Date' column to datetime format for proper plotting
filtered_data['Date'] = pd.to_datetime(filtered_data['Date'], errors='coerce')

# Set the 'Date' column as the index
filtered_data.set_index('Date', inplace=True)

# Plotting only the first four fuel types over time
plt.figure(figsize=(14, 8))
for column in ['RON 95-III', 'E5 RON 92-II', 'DO 0,05S-II', 'KO']:
    if column in filtered_data.columns:
        plt.plot(filtered_data.index, filtered_data[column], label=column)

# Adding plot details
plt.title("Price Changes of Selected Fuel Types Over Time")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend(title="Fuel Types")
plt.grid(True)
plt.tight_layout()
plt.show()
```
