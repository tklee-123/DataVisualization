```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fast_charge = pd.read_csv("data/charging_points/fast_charging_countries.csv")
slow_charge = pd.read_csv("data/charging_points/slow_charging_countries.csv")
```

```python
# Stacked bar plot
plt.figure(figsize=(14, 7))

countries = ["China", "Europe", "United States", "Other countries"]
df_plot = slow_charge.set_index("Year")
ax = df_plot[countries].plot(kind="bar", stacked=True, colormap="Set3")
ax.set_xlabel("Year")
ax.set_ylabel("The number of Standard Charging points (Million)")
ax.set_title("Electric Car Market Share by Region")
ax.legend(title="Region")
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Stacked bar plot
plt.figure(figsize=(14, 7))

countries = ["China", "Europe", "United States", "Other countries"]
df_plot = fast_charge.set_index("Year")
ax = df_plot[countries].plot(kind="bar", stacked=True, colormap="coolwarm")
ax.set_xlabel("Year")
ax.set_ylabel("The number of Fast Charging points (Million)")
ax.set_title("Electric Car Market Share by Region")
ax.legend(title="Region")
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
